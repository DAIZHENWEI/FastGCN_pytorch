```python
import numpy as np
import matplotlib.pyplot as plt
```

## Compare sampling methods

```python
Sage = np.load('GraphSage_accuracy_list_pubmed_batchsize183.npy')
Sage_random = np.load('GraphSage_accuracy_list_pubmed_random_ratio0.01.npy')
Sage_laplacian = np.load('GraphSage_accuracy_list_pubmed_laplacian_ratio0.01.npy')
# Sage_degree_10 = np.load('GraphSage_accuracy_list_pubmed_degree_ratio0.01.npy')
Sage_kmedoid = np.load('GraphSage_accuracy_list_pubmed_Kmedoid_ratio0.01.npy')
num_batch = int(18217/183+1)
```

```python
plt.plot(np.arange(1, len(Sage)+1)*num_batch, Sage, linestyle = '--', label = 'GraphSage')
plt.plot(np.arange(1, len(Sage_random)+1)*num_batch*0.01, Sage_random, linestyle = '-.', label = 'Random')
plt.plot(np.arange(1, len(Sage_laplacian)+1)*num_batch*0.01, Sage_laplacian, linestyle = ':', label = 'Laplacian')
plt.plot(np.arange(1, len(Sage_kmedoid)+1)*num_batch*0.01, Sage_kmedoid, linestyle = '-', label = 'Kmedoid')

plt.title('Pubmed, ratio = 0.01', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
# plt.ylim([0.4, 0.95])
plt.xlim([-20,3000])
# plt.savefig('../Figure/Reddit_accuracy_active_sampling_method_comparison_ratio_0.2.png')
plt.show()
```

```python
Sage = np.load('GraphSage_accuracy_list_pubmed_batchsize183.npy')
Sage_random = np.load('GraphSage_accuracy_list_pubmed_random_ratio0.1.npy')
Sage_laplacian = np.load('GraphSage_accuracy_list_pubmed_laplacian_ratio0.1.npy')
Sage_degree = np.load('GraphSage_accuracy_list_pubmed_degree_ratio0.1.npy')
Sage_kmedoid = np.load('GraphSage_accuracy_list_pubmed_Kmedoid_ratio0.1.npy')

num_batch = int(18217/256+1)
```

## Cross Entropy

```python
Sage = np.load('GraphSage_accuracy_list_pubmed.npy')
Sage_laplacian_05 = np.load('GraphSage_accuracy_list_pubmed_laplacian_ratio0.1.npy')
Sage_uncertain_05 = np.load('GraphSage_accuracy_list_pubmed_active_random_uncertain_epochs600_pretrain100_ratio0.1_sample_freq20.npy')
Sage_uncertain_incre_05 = np.load('GraphSage_accuracy_list_pubmed_active_random_uncertain_increment_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
Sage_uncertain_kick_05 = np.load('GraphSage_accuracy_list_pubmed_active_random_uncertain_kick_out_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')

num_batch = int(18217/256+1)

samples_count = np.load('Sample_count_list_pubmed_active_random_uncertain_increment_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
samples_cumsum = np.cumsum(samples_count)
num_batch_uncertain = [x/256 for x in samples_cumsum]

samples_count_kick = np.load('Sample_count_list_pubmed_active_random_uncertain_kick_out_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
samples_cumsum_kick = np.cumsum(samples_count_kick)
num_batch_uncertain_kick = [x/256 for x in samples_cumsum_kick]
```

```python
plt.plot(np.arange(1, len(Sage)+1)*num_batch, Sage, label = 'Standard')
# plt.plot(np.arange(1, len(Sage_random_05)+1)*int(num_batch*0.05), Sage_random_05, linestyle = '-.', label = 'Random')
# plt.plot(np.arange(1, len(Sage_laplacian_05)+1)*num_batch*0.1, Sage_laplacian_05, linestyle = '--', label = 'Laplacian, 5%')
plt.plot(np.arange(1, len(Sage_uncertain_05)+1)*num_batch*0.1, Sage_uncertain_05, linestyle = '-.', label = 'CE, pretrain 5%')
plt.plot(num_batch_uncertain, Sage_uncertain_incre_05, linestyle = '-.', label = 'CE incre, pretrain 5%')
plt.plot(num_batch_uncertain_kick, Sage_uncertain_kick_05, linestyle = ':', label = 'CE kick out, pretrain 5%')

plt.title('Pubmed, entropy based, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.4, 0.93])
plt.xlim([-100, 5000])
# plt.xscale('log')
# plt.savefig('../Figure/Pubmed_accuracy_active_uncertainty_matching_comparison_ratio_0.05.png')
plt.show()
```

## EL2N

```python
Sage = np.load('GraphSage_accuracy_list_pubmed.npy')
Sage_random_05 = np.load('GraphSage_accuracy_list_pubmed_laplacian_ratio0.1.npy')
Sage_EL2N_incre_05 = np.load('GraphSage_accuracy_list_pubmed_active_random_EL2N_increment_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
Sage_EL2N_kick_05 = np.load('GraphSage_accuracy_list_pubmed_active_random_EL2N_kick_out_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')

num_batch = int(18217/256+1)

samples_count = np.load('Sample_count_list_pubmed_active_random_EL2N_increment_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
samples_cumsum = np.cumsum(samples_count)
num_batch_EL2N = [x/256 for x in samples_cumsum]

samples_count_kick = np.load('Sample_count_list_pubmed_active_random_EL2N_kick_out_pretrain100_epochs600_pretrain_ratio0.05_ratio0.01_sample_freq20.npy')
samples_cumsum_kick = np.cumsum(samples_count_kick)
num_batch_EL2N_kick = [x/256 for x in samples_cumsum_kick]
```

```python
plt.plot(np.arange(1, len(Sage)+1)*num_batch, Sage, label = 'Standard')
plt.plot(np.arange(1, len(Sage_random_05)+1)*num_batch*0.1, Sage_random_05, linestyle = '--', label = 'Random, 5%')
# plt.plot(np.arange(1, len(Sage_EL2N_05)+1)*num_batch*0.1, Sage_EL2N_05, linestyle = '-.', label = 'CE, pretrain 5%')
plt.plot(num_batch_EL2N, Sage_EL2N_incre_05, linestyle = '-.', label = 'EL2N incre, pretrain 5%')
plt.plot(num_batch_EL2N_kick, Sage_EL2N_kick_05, linestyle = ':', label = 'EL2N kick, pretrain 5%')

plt.title('Pubmed, EL2N, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.4, 0.93])
plt.xlim([-100, 5000])
# plt.xscale('log')
# plt.savefig('../Figure/Pubmed_accuracy_active_EL2Nty_matching_comparison_ratio_0.05.png')
plt.show()
```
