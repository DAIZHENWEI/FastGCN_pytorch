```python
import numpy as np
import matplotlib.pyplot as plt
```

## Embeddings clustering

```python
LADIES = np.load('test_accuracy_list_pubmed_LADIES.npy')
LADIES_laplacian_05 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.05.npy')
LADIES_embed_05 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_embed_cluster_compround800_pretrain200_ratio0.05.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES))*num_batch, LADIES, label = 'Standard')
plt.plot(np.arange(len(LADIES_laplacian_05))*num_batch*0.05, LADIES_laplacian_05, linestyle = '--', label = 'Laplacian, 5%')
plt.plot(np.arange(len(LADIES_embed_05))*num_batch*0.05, LADIES_embed_05, linestyle = '-.', label = 'Embed Cluster, 5%')

plt.title('Pubmed, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.7, 0.9])
plt.xlim([100, 2000])
plt.savefig('../Figure/Pubmed_accuracy_active_embed_cluster_comparison_ratio_0.05.png')
plt.show()
```
