### Evaluate performance

```python
ratio = np.array([0.03, 0.04, 0.05, 0.07, 0.10])

Random_acc = np.array([corafull_random_acc_03, corafull_random_acc_04, corafull_random_acc_05, corafull_random_acc_07, 
                       corafull_random_acc_10])
Laplacian_acc = np.array([corafull_laplacian_acc_03, corafull_laplacian_acc_04, corafull_laplacian_acc_05, 
                          corafull_laplacian_acc_07, corafull_laplacian_acc_10])
FeatProp_acc = np.array([corafull_kmedoid_acc_03, corafull_kmedoid_acc_04, corafull_kmedoid_acc_05, 
                         corafull_kmedoid_acc_07, corafull_kmedoid_acc_10])
Grain_ball_acc = np.array([corafull_Grain_ball_acc_03, corafull_Grain_ball_acc_04, corafull_Grain_ball_acc_05, 
                           corafull_Grain_ball_acc_07, corafull_Grain_ball_acc_10])
Grain_NND_acc = np.array([corafull_Grain_NND_acc_03, corafull_Grain_NND_acc_04, corafull_Grain_NND_acc_05, 
                          corafull_Grain_NND_acc_07, corafull_Grain_NND_acc_10])
GIDUS_acc = np.array([corafull_batch_at_scale_acc_03, corafull_batch_at_scale_acc_04, corafull_batch_at_scale_acc_05, 
                      corafull_batch_at_scale_acc_07, corafull_batch_at_scale_acc_10])
AGE_acc = np.array([corafull_Grain_AGE_03, corafull_Grain_AGE_04, corafull_Grain_AGE_05, corafull_Grain_AGE_07,
                    corafull_Grain_AGE_10])
ANRMAB_acc = np.array([corafull_ANRMAB_03, corafull_ANRMAB_04, corafull_ANRMAB_05, corafull_ANRMAB_07, corafull_ANRMAB_10])

index = [0, 1, 2, 3, 4]

plt.plot(ratio[index], GIDUS_acc[index], label="ScatterSample", marker = "o", markersize=5)
plt.plot(ratio[index], AGE_acc, label="AGE", marker = "o", markersize=5)
plt.plot(ratio[index], ANRMAB_acc, label="ANRMAB", marker = "o", markersize=5)
plt.plot(ratio[index], Random_acc[index], label="Random", marker = "o", markersize=5)
plt.plot(ratio[index], FeatProp_acc[index], label="FeatProp", marker = "o", markersize=5)
# plt.plot(ratio[index], Laplacian_acc[index], label="Laplacian", marker = "o", markersize=5)
plt.plot(ratio[index], Grain_ball_acc[index], label="Grain(ball-D)", marker = "o", markersize=5)
plt.plot(ratio[index], Grain_NND_acc[index], label="Grain(NND)", marker = "o", markersize=5)
plt.xlabel('Sampling Ratio', fontsize = 14)
plt.ylabel('Test Accuracy', fontsize = 14)
plt.xticks( fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.tight_layout()
plt.savefig("../Figure/corafull_accuracy_comparison.png")
plt.show()
```
