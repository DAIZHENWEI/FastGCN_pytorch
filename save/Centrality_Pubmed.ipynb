```python
import numpy as np
import matplotlib.pyplot as plt
```

## Compare sampling methods

```python
Random = np.load('test_accuracy_list_pubmed_Random.npy')
Random_random_10 = np.load('test_accuracy_list_pubmed_Random_active_random_ratio0.1.npy')
Random_laplacian_10 = np.load('test_accuracy_list_pubmed_Random_active_laplacian_ratio0.1.npy')
Random_degree_10 = np.load('test_accuracy_list_pubmed_Random_active_degree_ratio0.1.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(1, len(Random)+1)*num_batch, Random, linestyle = '--', label = 'Standard')
plt.plot(np.arange(1, len(Random_random_10)+1)*int(num_batch*0.1), Random_random_10, linestyle = '-.', label = 'Random')
plt.plot(np.arange(1, len(Random_laplacian_10)+1)*int(num_batch*0.1), Random_laplacian_10, linestyle = ':', label = 'Laplacian')
plt.plot(np.arange(1, len(Random_degree_10)+1)*int(num_batch*0.1), Random_degree_10, linestyle = '-', label = 'Degree')

plt.title('Pubmed,joint sample neighbors, ratio = 0.1', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.6, 0.9])
plt.xlim([-50,2000])
# plt.savefig('../Figure/Pubmed_accuracy_active_sampling_method_comparison_ratio_0.2.png')
plt.show()
```

```python
LADIES = np.load('test_accuracy_list_pubmed_LADIES.npy')
LADIES_random_10 = np.load('test_accuracy_list_pubmed_LADIES_active_random_ratio0.1.npy')
LADIES_laplacian_10 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.1.npy')
LADIES_degree_10 = np.load('test_accuracy_list_pubmed_LADIES_active_degree_ratio0.1.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(1, len(LADIES)+1)*num_batch, LADIES, linestyle = '--', label = 'Standard')
plt.plot(np.arange(1, len(LADIES_random_10)+1)*int(num_batch*0.1), LADIES_random_10, linestyle = '-.', label = 'Random')
plt.plot(np.arange(1, len(LADIES_laplacian_10)+1)*int(num_batch*0.1), LADIES_laplacian_10, linestyle = ':', label = 'Laplacian')
plt.plot(np.arange(1, len(LADIES_degree_10)+1)*int(num_batch*0.1), LADIES_degree_10, linestyle = '-', label = 'Degree')

plt.title('Pubmed, ratio = 0.1', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.6, 0.9])
plt.xlim([-50,2000])
# plt.savefig('../Figure/Pubmed_accuracy_active_sampling_method_comparison_ratio_0.2.png')
plt.show()
```
