```python
import numpy as np
import matplotlib.pyplot as plt
```

## Compare sampling methods

```python
Random = np.load('test_accuracy_list_ogbn_arxiv_Random.npy')[0:25]
Random_random_10 = np.load('test_accuracy_list_ogbn_arxiv_Random_active_random_ratio0.05.npy')
Random_laplacian_10 = np.load('test_accuracy_list_ogbn_arxiv_Random_active_laplacian_ratio0.05.npy')
Random_degree_10 = np.load('test_accuracy_list_ogbn_arxiv_Random_active_degree_ratio0.05.npy')
num_batch = int(90941/256+1)
```

```python
plt.plot(np.arange(1, len(Random)+1)*num_batch, Random, linestyle = '--', label = 'Standard')
plt.plot(np.arange(1, len(Random_random_10)+1)*int(num_batch*0.05), Random_random_10, linestyle = '-.', label = 'Random')
plt.plot(np.arange(1, len(Random_laplacian_10)+1)*int(num_batch*0.05), Random_laplacian_10, linestyle = ':', label = 'Laplacian')
plt.plot(np.arange(1, len(Random_degree_10)+1)*int(num_batch*0.05), Random_degree_10, linestyle = '-', label = 'Degree')

plt.title('ogbn-arxiv, joint sample neighbors, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.2, 0.65])
plt.xlim([-200,5000])
# plt.savefig('../Figure/Reddit_accuracy_active_sampling_method_comparison_ratio_0.2.png')
plt.show()
```
