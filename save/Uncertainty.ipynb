```python
import numpy as np
import matplotlib.pyplot as plt
```

## Uncertainty measure

```python
LADIES = np.load('test_accuracy_list_ogbn_arxiv_LADIES.npy')[0:25]
LADIES_laplacian_05 = np.load('test_accuracy_list_ogbn_arxiv_LADIES_active_laplacian_ratio0.05.npy')
LADIES_uncertain_05 = np.load('test_accuracy_list_ogbn_arxiv_LADIES_active_laplacian_uncertain_compround800_pretrain20_ratio0.05.npy')
num_batch = int(90941/256+1)
```

```python
plt.plot(np.arange(1, len(LADIES)+1)*num_batch, LADIES, label = 'Standard')
# plt.plot(np.arange(1, len(LADIES_random_05)+1)*int(num_batch*0.05), LADIES_random_05, linestyle = '-.', label = 'Random')
plt.plot(np.arange(1, len(LADIES_laplacian_05)+1)*num_batch*0.05, LADIES_laplacian_05, linestyle = '--', label = 'Laplacian, 5%')
plt.plot(np.arange(1, len(LADIES_uncertain_05)+1)*num_batch*0.05, LADIES_uncertain_05, linestyle = '-.', label = 'Entropy based, 5%')

plt.title('ogbn_arxiv, entropy based, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.2, 0.65])
plt.xlim([-100, 5000])
# plt.xscale('log')
# plt.savefig('../Figure/Pubmed_accuracy_active_uncertainty_matching_comparison_ratio_0.05.png')
plt.show()
```

## Effect of pre-training epochs

```python
LADIES_uncertain_05_0 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain0_ratio0.05.npy')
LADIES_uncertain_05_50 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_compround800_pretrain50_ratio0.05.npy')
LADIES_uncertain_05_100 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain100_ratio0.05.npy')
LADIES_uncertain_05_200 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain200_ratio0.05.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES_uncertain_05_0))*num_batch*0.05, LADIES_uncertain_05_0, linestyle = '--', label = '5%, 0 epochs')
plt.plot(np.arange(len(LADIES_uncertain_05_50))*num_batch*0.05, LADIES_uncertain_05_50, label = '5%, 50 epochs')
plt.plot(np.arange(len(LADIES_uncertain_05_100))*num_batch*0.05, LADIES_uncertain_05_100, linestyle = '-.', label = '5%, 100 epochs')
plt.plot(np.arange(len(LADIES_uncertain_05_200))*num_batch*0.05, LADIES_uncertain_05_200, linestyle = '-', label = '5%, 200 epochs')

plt.title('Pubmed, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.6, 0.9])
# plt.savefig('../Figure/Pubmed_accuracy_active_uncertainty_matching_comparison_ratio_0.05.png')
plt.show()
```

```python
LADIES_uncertain_01_0 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain0_ratio0.01.npy')
LADIES_uncertain_01_100 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain100_ratio0.01.npy')
LADIES_uncertain_01_200 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain200_ratio0.01.npy')
LADIES_uncertain_01_300 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_pretrain300_ratio0.01.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES_uncertain_01_0))*1, LADIES_uncertain_01_0, linestyle = '--', label = '1%, 0 epochs')
plt.plot(np.arange(len(LADIES_uncertain_01_100))*1, LADIES_uncertain_01_100, linestyle = '-.', label = '1%, 100 epochs')
plt.plot(np.arange(len(LADIES_uncertain_01_200))*1, LADIES_uncertain_01_200, linestyle = '-', label = '1%, 200 epochs')
plt.plot(np.arange(len(LADIES_uncertain_01_300))*1, LADIES_uncertain_01_300, linestyle = ':', label = '1%, 300 epochs')

plt.title('Pubmed, ratio = 0.01', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.6, 0.9])
# plt.savefig('../Figure/Pubmed_accuracy_active_uncertainty_matching_comparison_ratio_0.01.png')
plt.show()
```
