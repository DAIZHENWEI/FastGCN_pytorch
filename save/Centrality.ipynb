```python
import numpy as np
import matplotlib.pyplot as plt
```

## Compare sampling ratios

```python
LADIES = np.load('test_accuracy_list_reddit_LADIES.npy')[0:25]
LADIES_20 = np.load('test_accuracy_list_reddit_LADIES_active_laplacian_ratio0.2.npy')[0:100]
LADIES_10 = np.load('test_accuracy_list_reddit_LADIES_active_laplacian_ratio0.1.npy')
LADIES_5 = np.load('test_accuracy_list_reddit_LADIES_active_laplacian_ratio0.05.npy')
num_batch = int(153431/256+1)
```

```python
plt.plot(np.arange(1, len(LADIES)+1)*num_batch, LADIES, linestyle = '--', label = 'Standard')
plt.plot(np.arange(1, len(LADIES_20)+1)*int(num_batch*0.2), LADIES_20, linestyle = ':', label = 'Laplacian, ratio=0.2')
plt.plot(np.arange(1, len(LADIES_10)+1)*int(num_batch*0.1), LADIES_10, linestyle = '-', label = 'Laplacian, ratio=0.1')
plt.plot(np.arange(1, len(LADIES_5)+1)*int(num_batch*0.05), LADIES_5, label = 'Laplacian, ratio=0.05')

plt.ylim([0.7, 0.93])
plt.xlim([-100,10000])
plt.title('Reddit, Laplacian Sampling', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
# plt.savefig('../Figure/Pubmed_accuracy_active_sampling_ratio_comparison_laplacian.png')
plt.show()
```

```python
LADIES = np.load('test_accuracy_list_pubmed_LADIES.npy')[0:300]
LADIES_50 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.5.npy')
LADIES_20 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.2.npy')
LADIES_10 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.1.npy')
LADIES_5 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.05.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES))*num_batch, LADIES, linestyle = '--', label = 'Standard')
plt.plot(np.arange(len(LADIES_50))*int(num_batch*0.5), LADIES_50, linestyle = '-.', label = 'Laplacian, ratio=0.5')
plt.plot(np.arange(len(LADIES_20))*int(num_batch*0.2), LADIES_20, linestyle = ':', label = 'Laplacian, ratio=0.2')
plt.plot(np.arange(len(LADIES_10))*int(num_batch*0.1), LADIES_10, linestyle = '-', label = 'Laplacian, ratio=0.1')
plt.plot(np.arange(len(LADIES_5))*int(num_batch*0.05), LADIES_5, label = 'Laplacian, ratio=0.05')

plt.ylim([0.7, 0.92])
plt.xlim([-100,3000])
plt.title('Pubmed, Laplacian Sampling', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.savefig('../Figure/Pubmed_accuracy_active_sampling_ratio_comparison_laplacian.png')
plt.show()
```

### Incorporate neighbor elimination

```python
LADIES_laplacian_20 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.2.npy')
LADIES_degree_20 = np.load('test_accuracy_list_pubmed_LADIES_active_degree_ratio0.2.npy')
LADIES_laplacian_cluster_20 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_cluster_ratio0.2.npy')
LADIES_degree_cluster_20 = np.load('test_accuracy_list_pubmed_LADIES_active_degree_cluster_ratio0.2.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES_laplacian_20))*int(num_batch*0.2), LADIES_laplacian_20, linestyle = ':', color = 'blue', label = 'Laplacian')
plt.plot(np.arange(len(LADIES_degree_20))*int(num_batch*0.2), LADIES_degree_20, linestyle = ':', color ='orange',  label = 'Degree')
plt.plot(np.arange(len(LADIES_laplacian_cluster_20))*int(num_batch*0.2), LADIES_laplacian_cluster_20, color = 'blue', label = 'Laplacian, exclude neighbors')
plt.plot(np.arange(len(LADIES_degree_cluster_20))*int(num_batch*0.2), LADIES_degree_cluster_20, color='orange', label = 'Degree, exclude neighbors')

plt.title('Pubmed, ratio = 0.2', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.7, 0.9])
plt.xlim([-300,9000])
plt.savefig('../Figure/Pubmed_accuracy_active_cluster_vs_no_cluster_ratio_0.2.png')
plt.show()
```
