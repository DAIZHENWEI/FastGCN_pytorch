## Compare sampling ratios

```python
LADIES = np.load('test_accuracy_list_pubmed_LADIES.npy')
LADIES_50 = np.load('test_accuracy_list_pubmed_LADIES_active_ratio0.5.npy')
LADIES_20 = np.load('test_accuracy_list_pubmed_LADIES_active_ratio0.2.npy')
LADIES_10 = np.load('test_accuracy_list_pubmed_LADIES_active_ratio0.1.npy')
LADIES_5 = np.load('test_accuracy_list_pubmed_LADIES_active_ratio0.05.npy')
num_batch = int(18717/256+1)
```

```python
np.max(LADIES_5), np.max(LADIES_10), np.max(LADIES_20)
```

```python
plt.plot(np.arange(len(LADIES))*num_batch, LADIES, linestyle = '--', label = 'LADIES')
plt.plot(np.arange(len(LADIES_50))*int(num_batch*0.5), LADIES_50, linestyle = '-.', label = 'LADIES, ratio=0.5')
plt.plot(np.arange(len(LADIES_20))*int(num_batch*0.2), LADIES_20, linestyle = ':', label = 'LADIES, ratio=0.2')
plt.plot(np.arange(len(LADIES_10))*int(num_batch*0.1), LADIES_10, linestyle = '-', label = 'LADIES, ratio=0.1')
plt.plot(np.arange(len(LADIES_5))*int(num_batch*0.05), LADIES_5, label = 'LADIES, ratio=0.05')

plt.xlim([-300,10000])
plt.title('Pubmed, Laplacian Sampling', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.savefig('../Figure/Pubmed_accuracy_active_sampling_ratio_comparison_laplacian.png')
plt.show()
```

### Incorporate neighbor elimination

```python
LADIES_laplacian_20 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.2.npy')
LADIES_degree_20 = np.load('test_accuracy_list_pubmed_LADIES_active_degree_ratio0.2.npy')
LADIES_laplacian_cluster_20 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_cluster_ratio0.2.npy')
LADIES_degree_cluster_20 = np.load('test_accuracy_list_pubmed_LADIES_active_degree_cluster_ratio0.2.npy')
num_batch = int(18717/256+1)
```

```python
plt.plot(np.arange(len(LADIES_laplacian_20))*int(num_batch*0.2), LADIES_laplacian_20, linestyle = ':', color = 'blue', label = 'Laplacian')
plt.plot(np.arange(len(LADIES_degree_20))*int(num_batch*0.2), LADIES_degree_20, linestyle = ':', color ='orange',  label = 'Degree')
plt.plot(np.arange(len(LADIES_laplacian_cluster_20))*int(num_batch*0.2), LADIES_laplacian_cluster_20, color = 'blue', label = 'Laplacian, exclude neighbors')
plt.plot(np.arange(len(LADIES_degree_cluster_20))*int(num_batch*0.2), LADIES_degree_cluster_20, color='orange', label = 'Degree, exclude neighbors')

plt.title('Pubmed, ratio = 0.2', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.7, 0.9])
plt.xlim([-300,9000])
plt.savefig('../Figure/Pubmed_accuracy_active_cluster_vs_no_cluster_ratio_0.2.png')
plt.show()
```

## Uncertainty measure

```python
LADIES = np.load('test_accuracy_list_pubmed_LADIES.npy')
LADIES_laplacian_05 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_ratio0.05.npy')
LADIES_uncertain_05 = np.load('test_accuracy_list_pubmed_LADIES_active_laplacian_uncertain_ratio0.05.npy')
num_batch = int(18717/256+1)
```

```python
np.max(LADIES_uncertain_05[0:200]), np.max(LADIES_uncertain_05[200:])
```

```python
plt.plot(np.arange(len(LADIES))*num_batch, LADIES, label = 'Standard')
plt.plot(np.arange(len(LADIES_laplacian_05))*int(num_batch*0.05), LADIES_laplacian_05, linestyle = '--', label = 'Laplacian, 5%')
plt.plot(np.arange(len(LADIES_uncertain_05))*int(num_batch*0.05), LADIES_uncertain_05, linestyle = '-.', label = 'Uncertainty, 5%')

plt.title('Pubmed, ratio = 0.05', fontsize = 16)
plt.xlabel('Number of Iterations', fontsize = 14)
plt.ylabel('Accuracy', fontsize = 14)
plt.legend(fontsize = 14)
plt.ylim([0.7, 0.9])
plt.xlim([10, 50000])
plt.xscale('log')
plt.savefig('../Figure/Pubmed_accuracy_active_uncertainty_matching_comparison_ratio_0.05.png')
plt.show()
```
